```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
matrix = np.array([[1.4, 0.7, 2.9, 8.5, 7.2],
                   [1.1, 1.5, 4.7, 2.9, 5.1],
                   [2.8, 2.9, 6.2, 7.8, 8.8],
                   [2.5, 6.8, 1.3, 0, 4],
                   [2.3, 5.0, 2.7, 8.0, 9.0]])
```

```python
# a) Tính định thức của ma trận A
print("Định thức của ma trận A:", np.linalg.det(matrix))
```

```python
# b> Tính ma trận nghịch đảo của A
print("Ma trận nghịch đảo của A:")
np.linalg.inv(matrix)
```

```python
# c) Tính các giá trị riêng của A
values, vectors = np.linalg.eig(matrix)
print("Giá trị riêng của ma trận A:", values)
```

```python
# d) Tính vecto riêng của từng giá trị riêng
for i in range(len(values)):
    print("Giá trị riêng:", values[i], ". Vecto riêng tương ứng:", vectors[i])
```

```python
# e) Tính tichs của 2 ma trận A1 x A2 trong đó A1 gồm các hàng 1, 3, 4, 5, cột 1, 2, 3, 5 của A. A2 là các hàng 2, 3, 4, 5, cột 1, 2, 3, 4 của A
matrix_a1 = matrix[[0, 2, 3, 4]][:, [0, 1, 2, 4]]
matrix_a2 = matrix[[1, 2, 3, 4]][:, [0, 1, 2, 3]]
```

```python
print("A1 x A2:")
matrix_a1.dot(matrix_a2)
```

```python
# f) Cho B: bij = aij nếu aij >=4; bij = 0 nếu aij < 4. Tính det(B) rank(B)
matrix_B = matrix.copy()
matrix_B[matrix < 4] = 0
print("Định thức ma trận B:", np.linalg.det(matrix_B))
print("Hạng ma trận B:", np.linalg.matrix_rank(matrix_B))
```

```python
df = pd.read_csv('Example1.csv')
df
```

```python
# b) Chuẩn hóa cột A, C, E theo min, max
df['A'] = (df['A'] - min(df['A'])) / (max(df['A']) - min(df['A']))
df['C'] = (df['C'] - min(df['C'])) / (max(df['C']) - min(df['C']))
df['E'] = (df['E'] - min(df['E'])) / (max(df['E']) - min(df['E']))
df
```

```python
# c) Chuẩn hóa cột B, D, F, G theo Z-score
df['B'] = (df['B'] - np.average(df['B'])) / np.std(df['B'])
df['D'] = (df['D'] - np.average(df['D'])) / np.std(df['D'])
df['F'] = (df['F'] - np.average(df['F'])) / np.std(df['F'])
df['G'] = (df['G'] - np.average(df['G'])) / np.std(df['G'])
df
```

```python
# d) Chuẩn hóa các cột A, B, C, D, E, F, G theo min-max về đoạn [0.2, 0.7]
# for col in ['A', 'B', 'C', 'D', 'E', 'F', 'G']:
#     df[col] = (0.7 - 0.2) * (df[col] - min(df[col])) / (max(df[col]) - min(df[col])) + 0.2
# df

scaler = MinMaxScaler(feature_range=(0.2, 0.7))
A_to_G = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
df[A_to_G] = pd.DataFrame(scaler.fit_transform(df[A_to_G]), columns=df[A_to_G].columns)
df
```

```python
# e) Chuẩn hóa cột H về dạng số
label_encoder = LabelEncoder()
df['H'] = label_encoder.fit_transform(df['H'])
df
```
